Let me help you understand how to explore package contents in Python. There are several ways to inspect what's inside a package like yfinance:

1. Using `dir()` to see all attributes and methods:
```python
import yfinance as yf
print(dir(yf))
```

2. Using `help()` to get detailed documentation:
```python
help(yf)
# Or for a specific class/function
help(yf.Ticker)
```

3. To see the actual source code, you can:

   a. Use Python's `inspect` module:
   ```python
   import inspect
   print(inspect.getsource(yf.Ticker))
   ```

   b. Find the package location and view the files directly:
   ```python
   import yfinance
   print(yfinance.__file__)  # This shows where the package is installed
   ```

4. You can also explore the package's GitHub repository. For yfinance, it's at: https://github.com/ranaroussi/yfinance

Here's an example of inspecting the yfinance Ticker class structure:

```python
# Get all methods of the Ticker class
ticker = yf.Ticker('GOOGL')
methods = [method for method in dir(ticker) if not method.startswith('_')]
print("Available methods:", methods)

# Get info about a specific method
help(ticker.history)
```

The yfinance package is organized roughly like this:
- `Ticker` class: Main class for getting stock information
  - `history()`: Gets historical market data
  - `info`: Gets stock information
  - `actions`: Gets dividend and split data
  - `financials`: Gets financial statements
  - And many more methods

When you want to understand a package's internals, I recommend this process:
1. Start with `dir()` to see what's available
2. Use `help()` for documentation
3. Use the package's documentation website
4. Check the GitHub repository for source code
5. Use `inspect` module for deeper exploration
